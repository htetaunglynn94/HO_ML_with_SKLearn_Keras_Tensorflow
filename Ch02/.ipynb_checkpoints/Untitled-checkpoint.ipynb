```python
import tarfile
import urllib.request

import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
# GitHub raw file URL
dl_path = "https://github.com/htetaunglynn94/HO_ML_with_SKLearn_Keras_Tensorflow/raw/main/datasets/housing.tgz"

# Download the file
urllib.request.urlretrieve(dl_path, "housing.tgz")

# Extract the file
with tarfile.open("housing.tgz", "r:gz") as housing_tgz:
    housing_tgz.extractall()

```
